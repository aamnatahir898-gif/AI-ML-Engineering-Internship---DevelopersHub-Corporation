```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, confusion_matrix
```

***load dataset***

```python
import pandas as pd

# Load dataset using pandas
url = "https://raw.githubusercontent.com/mwaskom/seaborn-data/master/iris.csv"
df = pd.read_csv(url)

# Display first 5 rows
df.head()
```

```python
# Shape of dataset
print("Shape:", df.shape)

# Column names
print("Columns:", df.columns)

# First few rows
df.head()
```

```python
# Dataset information
df.info()
```

```python
# Summary statistics
df.describe()
```

**VISUALIZATIONS**


***scatterplot***

```python
plt.figure()
sns.scatterplot(x='sepal_length', y='sepal_width', hue='species', data=df)
plt.title("Sepal Length vs Sepal Width")
plt.show()
# This scatter plot shows the relationship between sepal length and width.
# Different colors represent different species, helping us identify patterns.
```

***histogram***

```python
df.hist()
plt.suptitle("Histograms of Features")
plt.show()

# Histograms show how values of each feature are distributed.
```

***Box Plot***

```python
plt.figure()
sns.boxplot(data=df)
plt.title("Boxplot for Outliers Detection")
plt.show()


# Boxplot helps detect outliers and spread of data.
```

```python
# Convert categorical species into numbers
df['species'] = df['species'].astype('category').cat.codes

# Features and target
X = df.drop('species', axis=1)
y = df['species']
```

***Train-Test Split***

```python

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

***Model Training***

```python
model = LogisticRegression(max_iter=200, C=0.1)
model.fit(X_train, y_train)
```

***Prediction***

```python
y_pred = model.predict(X_test)
```

***Evaluation***

```python
# Accuracy
accuracy = accuracy_score(y_test, y_pred)
print("Accuracy:", accuracy)

# Confusion Matrix
print("Confusion Matrix:\n", confusion_matrix(y_test, y_pred))
cm = confusion_matrix(y_test, y_pred)
plt.figure(figsize=(4,3))
sns.heatmap(cm, annot=True, fmt='d', cmap='Blues')
plt.title('Confusion Matrix')
plt.ylabel('Actual')
plt.xlabel('Predicted')
plt.show()
```

 Results and Insights

1-The dataset contains 150 samples and 4 features.
2- No missing values were found.
3- Scatter plots show that some species are clearly separable.
4- Logistic Regression model achieved good accuracy.
5- The model successfully classifies different flower species.

Overall, the dataset is clean and suitable for machine learning tasks.
